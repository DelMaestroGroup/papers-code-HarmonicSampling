```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18,8))

#=============================================================================================
markerFC=['lightgray', 'khaki', 'gold', 'plum']
markerEC=['dimgray', 'darkkhaki', 'darkgoldenrod', 'violet']
makker=['^', 'v', '<', '>', 'h', 'H']

Ts = [1.0, 1.5, 2.0]

plt.subplot(2, 3, 1)
df_acc = np.loadtxt('../data/Fig_7/Low_Barrier_Height/StagingAcceptanceRatio_V0_p3.txt', delimiter=',', skiprows=1)
for i, T in enumerate(Ts):
    plt.plot(df_acc[:,0], df_acc[:, i+1], linestyle='-', marker=makker[i], markersize=10, color=markerFC[i], markerfacecolor=markerFC[i], markeredgecolor=markerEC[i], markeredgewidth=2, label=f'T = {T} K')# (P = {int(beads[i])})')

plt.subplot(2, 3, 4)
for i, T in enumerate(Ts):
    ECorr = np.loadtxt(f'../data/Fig_7/Low_Barrier_Height/EnergyAutocorrelation_V0_p3_T_{T}_K.txt', delimiter=',', skiprows=1)
    plt.errorbar(ECorr[:,0], ECorr[:, 1], yerr=ECorr[:, 2], linestyle='-', marker=makker[i], markersize=10, color=markerFC[i], markerfacecolor=markerFC[i], markeredgecolor=markerEC[i], markeredgewidth=2, ecolor = markerEC[i], capsize=5, label=f'T = {T} K')# (P = {int(beads[i])})')

#=============================================================================================
markerFC=['lightgray', 'khaki', 'gold', 'plum']
markerEC=['dimgray', 'darkkhaki', 'darkgoldenrod', 'violet']
makker=['^', 'v', '<', '>', 'h', 'H']

Ts = [1.5, 2.0, 2.5]

plt.subplot(2, 3, 2)
df_acc = np.loadtxt('../data/Fig_7/Intermediate_Barrier_Height/StagingAcceptanceRatio_V0_p6.txt', delimiter=',', skiprows=1)
for i, T in enumerate(Ts):
    plt.plot(df_acc[:,0], df_acc[:, i+1], linestyle='-', marker=makker[i], markersize=10, color=markerFC[i], markerfacecolor=markerFC[i], markeredgecolor=markerEC[i], markeredgewidth=2, label=f'T = {T} K')# (P = {int(beads[i])})')

plt.subplot(2, 3, 5)
for i, T in enumerate(Ts):
    ECorr = np.loadtxt(f'../data/Fig_7/Intermediate_Barrier_Height/EnergyAutocorrelation_V0_p6_T_{T}_K.txt', delimiter=',', skiprows=1)
    plt.errorbar(ECorr[:,0], ECorr[:, 1], yerr=ECorr[:, 2], linestyle='-', marker=makker[i], markersize=10, color=markerFC[i], markerfacecolor=markerFC[i], markeredgecolor=markerEC[i], markeredgewidth=2, ecolor = markerEC[i], capsize=5, label=f'T = {T} K')# (P = {int(beads[i])})')

#=============================================================================================
markerFC=['lightgray', 'khaki', 'gold', 'plum']
markerEC=['dimgray', 'darkkhaki', 'darkgoldenrod', 'violet']
makker=['^', 'v', '<', '>', 'h', 'H']

Ts = [2.5, 3.0, 4.0]

plt.subplot(2, 3, 3)
df_acc = np.loadtxt('../data/Fig_7/Large_Barrier_Height/StagingAcceptanceRatio_V0_1p.txt', delimiter=',', skiprows=1)
for i, T in enumerate(Ts):
    plt.plot(df_acc[:,0], df_acc[:, i+1], linestyle='-', marker=makker[i], markersize=10, color=markerFC[i], markerfacecolor=markerFC[i], markeredgecolor=markerEC[i], markeredgewidth=2, label=f'T = {T} K')# (P = {int(beads[i])})')

plt.subplot(2, 3, 6)
for i, T in enumerate(Ts):
    ECorr = np.loadtxt(f'../data/Fig_7/Large_Barrier_Height/EnergyAutocorrelation_V0_1p_T_{T}_K.txt', delimiter=',', skiprows=1)
    plt.errorbar(ECorr[:,0], ECorr[:, 1], yerr=ECorr[:, 2], linestyle='-', marker=makker[i], markersize=10, color=markerFC[i], markerfacecolor=markerFC[i], markeredgecolor=markerEC[i], markeredgewidth=2, ecolor = markerEC[i], capsize=5, label=f'T = {T} K')# (P = {int(beads[i])})')

#=============================================================================================
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.tick_params(labelsize=16)
    plt.grid()

for i in [4, 5, 6]:
    plt.subplot(2, 3, i)
    plt.xlabel(f'anharmonic contribution', fontsize=18)

for i in [1, 2, 3]:
    plt.subplot(2, 3, i)
    plt.ylabel(f'Acceptance ratio', fontsize=18)
    plt.subplot(2, 3, 3+i)
    plt.ylabel(f'Energy autocorrelation time', fontsize=18)

plt.subplot(2, 3, 1)
plt.title(f'Low barrier height', fontsize=16)
plt.yticks([0.5, 0.6, 0.7])
plt.legend(fontsize=14, loc=[0.6,0.1])
plt.subplot(2, 3, 2)
plt.title(f'Intermediate barrier height', fontsize=16)
plt.yticks([0.4, 0.5, 0.6])
plt.legend(fontsize=14, loc=[0.58,0.01])
plt.subplot(2, 3, 3)
plt.title(f'High barrier height', fontsize=16)
plt.yticks([0.4, 0.5, 0.6, 0.7])
plt.legend(fontsize=14, loc=[0.6,0.1])

plt.subplot(2, 3, 4)
plt.yticks([2, 4, 6])
plt.legend(fontsize=14, loc=[0.03,0.625])
plt.subplot(2, 3, 5)
plt.yticks([3, 4, 5])
plt.legend(fontsize=14, loc=[0.59,0.45])
plt.subplot(2, 3, 6)
plt.yticks([3, 4, 5])
plt.legend(fontsize=14, loc=[0.59,0.625])

plt.tight_layout()
plt.subplots_adjust(hspace=0)
plt.savefig('../figures/Fig7.pdf')
```
