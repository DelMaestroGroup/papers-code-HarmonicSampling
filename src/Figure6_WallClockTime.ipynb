```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wanhfree = np.loadtxt('../data/wanh_free_NT', skiprows=1,unpack=True)
wanhmixed = np.loadtxt('../data/wanh_mixed_accept_HD9',unpack=True)
wanhharmonic = np.loadtxt('../data/wanh_harmonic_NT', skiprows=1,unpack=True)
```

```python
manhfree = np.loadtxt('../data/manh_free_NT', skiprows=1,unpack=True)
manhmixed = np.loadtxt('../data/manh_mixed_accept_HD8',unpack=True)
manhharmonic = np.loadtxt('../data/manh_harmonic_NT', skiprows=1,unpack=True)
```

```python
sanhfree = np.loadtxt('../data/sanh_bench_free_acc', unpack=True)
sanhmixed = np.loadtxt('../data/sanh_bench_mixed_acc',unpack=True)
sanhharmonic = np.loadtxt('../data/sanh_bench_harm_acc', unpack=True)
```

```python
#x = np.array(['PIMC','M-PIMC','H-PIMC'])
#Get free time
wanhtime = np.array([wanhfree[-2][-1],wanhharmonic[-2][0],wanhmixed[-2][0]])
manhtime = np.array([manhfree[-2][-1],manhharmonic[-2][1],manhmixed[-2][1]])
sanhtime = np.array([sanhfree[-2][-1],sanhharmonic[-2][-1],sanhmixed[-2][-1]])
```

```python
plt.rc('xtick',labelsize=9)
plt.rc('ytick',labelsize=9)
fig,ax = plt.subplots(1,1,figsize=(2*3.4039, 2.10373),sharey=True,constrained_layout=True)
ax.set_ylabel('Time (s)', fontsize=14)
ax.bar([0.8,1.0,1.2],wanhtime,width=0.2,facecolor=['cornflowerblue','salmon','limegreen'], edgecolor=['royalblue','red','darkgreen'])
ax.set_yscale('log')
#ax.set_title('Weakly anharmonic', fontsize=18)
#ax2.set_ylabel('Time (s)', fontsize=8)
ax.set_yscale('log')
ax.bar([1.6,1.8,2.0],manhtime,width=0.2,facecolor=['cornflowerblue','salmon','limegreen'], edgecolor=['royalblue','red','darkgreen'])
#ax.set_title('Moderately anharmonic', fontsize=18)
#ax3.set_ylabel('Time (s)', fontsize=8)
ax.set_yscale('log')
ax.bar([2.4,2.6,2.8],sanhtime,width=0.2,facecolor=['cornflowerblue','salmon','limegreen'], edgecolor=['royalblue','red','darkgreen'])
ax.set_xticks([0.8,1.0,1.2,1.6,1.8,2.0,2.4,2.6,2.8],['PIMC','H-PIMC','M-PIMC','PIMC','H-PIMC','M-PIMC','PIMC','H-PIMC','M-PIMC'])
#ax.set_yticks([10**2,10**3,10**4])
ax.set_ylim(10,3*10**5/4)
ax.set_xlim(0.7,2.9)
ax.text(0.78,10**4/3.5,'Weakly Anharmonic',fontsize=10,bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.5))
ax.text(1.53,10**4/3.5,'Moderately Anharmonic',fontsize=10,bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.5))
ax.text(2.35,10**4/3.5,'Strongly Anharmonic',fontsize=10,bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.5))
ax.set_ylim(10,1.7*1000)
plt.savefig('../figures/wallclocktime.pdf')
```
