```python
import numpy as np
import emcee
```

```python
# THIS FUNCTION IS USED TO GET AVERAGE ENERGY AND STAGING ACCEPTANCE RATIO FOR ALL THE COMPUTATIONS

def getvals(df1):
    nbeads = list(set(df1[:,0]))
    nbeads.sort()
    priPIMC, priPIMC_std, acc_stage, acc_stage_std = [], [], [], []
    for nbead in nbeads:
        priPIMC_temp, acc_stage_temp = [], []
        for i in range(len(df1)):
            if df1[i, 0] == nbead:
                priPIMC_temp.append(df1[i, 1])
#                 acc_com_temp.append(df1[i, -2])
                acc_stage_temp.append(df1[i, -1])
        priPIMC.append(np.mean(priPIMC_temp))
        priPIMC_std.append(np.std(priPIMC_temp)/np.sqrt(len(priPIMC_temp)-1))
        acc_stage.append(np.mean(acc_stage_temp))
        acc_stage_std.append(np.std(acc_stage_temp)/np.sqrt(len(acc_stage_temp)-1))

    return nbeads, priPIMC, priPIMC_std, acc_stage, acc_stage_std

# 'output.txt' contains PIMC / HPIMC output of total energy, acceptance ratio as a function of beads number
df = np.loadtxt('output.txt', delimiter=',')
nbeads, priPIMC, priPIMC_std, acc_stage, acc_stage_std = getvals(df)
```
